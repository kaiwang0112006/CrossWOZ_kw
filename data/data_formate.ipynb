```python
import json
```

```python
with open(r"/mnt/datawow/kw/CrossWOZ/data/crosswoz/train.json") as f:
    data = json.load(f)
```

```python
len(data)
```

```python
data["391"]
```

```python
dataoutput = []
for k in data:
    conversation = []
    extractmsg = {}
    for msg in data[k]["messages"]:
        if "input" not in extractmsg:
            extractmsg["input"] = msg["content"]
        elif "output" not in extractmsg:
            extractmsg["output"] = msg["content"]
        if "input" in extractmsg and "output" in extractmsg:
            conversation.append(extractmsg)
            extractmsg = {}
    
    dataoutput.append({"conversation":conversation})
```

```python
print(json.dumps(dataoutput[:3], indent=4,ensure_ascii=False))
```
